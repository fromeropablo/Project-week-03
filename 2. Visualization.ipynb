###  2. The Zion Effect

If you don't like basketball, maybe you won't know anything about Zion Williamson. But I have to tell you something, since he arrrived in the NBA, he is genuinely a beast. Moreover, he's just 20 and he's already making history in the league.

```python
visual.zion_effect()
```

```python
zion_effect = nba_dataset.loc[(nba_dataset["PTS"] >= 25) & (nba_dataset["AGE"] < 26) & 
                (nba_dataset["GP"] >= 30) & (nba_dataset["REB"] >= 6) & 
                (nba_dataset["TS_PCT"] >= 0.60)]
plt.figure(figsize=(8,6))
sns.scatterplot(data=zion_effect, x="SEASON", y="AGE", hue="PLAYER_NAME")       
plt.tight_layout()
plt.show()
```

It could be a bit difficult to appreciate, but Zion Williamson is performing one of the best seasons for a 20-years player in NBA history. In the last 25 seasons, just six more players have averaged more than 25 points, 6 rebounds and a True Shooting Percentage above 60% with at least 30 games played. Only Zion have done that at such age. 

### 3. Who has been cooking at the Spanish League?

<img src="https://media.giphy.com/media/fZYlv7E1cvj6RExBEN/giphy.gif" width="750" align="center">

In Europe, most of the leagues of the continent have an stat column called PIR (Performance Index Rating). We are gonna find the best 10 players in ACB/Liga Endesa during the 21st Century. 

```python
acb_dataset["VAL"] = acb_dataset["VAL"].str.replace(",", ".").astype(float)
best_acb = acb_dataset.sort_values(by = ["VAL"], ascending = False)
best_acb = best_acb.head(10)



plt.figure(figsize=(16,9))
ax= sns.barplot(data = best_acb, x = "PLAYER_NAME", y = "VAL")

ax.set_xticklabels(ax.get_xticklabels(), rotation=25, ha="right")
plt.tight_layout()
plt.show()
```

Just three players from the Top 10 remains active and only two of them (Rudy Fernndez and Nikola Mirotic) are playing in the Spanish League in this season.
